```python
import folium

m = folium.Map(location = [33.37329056367668, 126.53190545103804])

tooltip = '클릭해보세요'

folium.Marker(
    [33.3617, 126.5292],
    popup = '<strong>한라산</strong>',
    tooltip = tooltip
).add_to(m)

folium.Marker(
    [33.5104, 126.4914],
    popup = '<strong>제주국제공항</strong>',
    icon = folium.Icon(color='#ff0000', icon='info-sign'),
    tooltip = tooltip
).add_to(m)

m
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import folium

df = pd.read_csv('./data/서귀포시오름현황.csv', encoding= 'cp949')
df
```

```python
df.iloc[0]
df.iloc[0]['오름명']
df.iloc[0]['위도']
df.iloc[0]['경도']
len(df)
```

```python
for i in range(len(df)):
  print(df.iloc[i]['오름명'],
        df.iloc[i]['위도'],
        df.iloc[i]['경도'])
```

```python
import folium
m = folium.Map(location = [33.37329056367668, 126.53190545103804])
# m = folium.Map(location = [33.3617, 126.5292])

tooltip = '클릭해보세요'

for i in range(len(df)):
  folium.Marker(
  [df.iloc[i]['위도'], df.iloc[i]['경도']],
  popup = folium.Popup(f'<strong>{df.iloc[i]["오름명"]}</strong>', max_width=500),
  tooltip = tooltip
  ).add_to(m)
m
```

```python
folium.Circle(
    [33.37329056367668, 126.53190545103804],
    radius = 10000,
    color = 'red',
    fill = False # True
).add_to(m)
m
```

```python
folium.PolyLine(
    locations = [[33.3617, 126.5292], [33.456482, 126.8727771]],
    tooltip = '한라산에서 대왕산까지!'
).add_to(m)
m
```
