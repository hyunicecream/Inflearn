## requets

- HTTP 요청을 보내는데 사용하는 라이브러리
- .text : str타입의 데이터를 return
- .headers : header(key/value 형식으로 데이터 저장)의 내용 확인
- .encoding : 인코딩 방식 확인
- .status_code : HTTP 요청에 대해서 요청이 성공했는지 실패했는지 혹은 어떤 상태인지 말해줌
- .ok : 데이터를 잘 불러오고 있는지 확인
