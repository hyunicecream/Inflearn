```python
# 문제 : 두 배열이 있을 때 출력물은?
# 1. (3, 2)
# 2. (2, 3)
# 3. (error) -> 정답
# 이유 : dot-product 연산을 하기 위해서는 한 배열의 행의 갯수와 연산을 할 배열의 갯수가 동일해야 한다. 
arr = np.array([[1, 2],
                [3, 4],
                [5, 6]])
arr2 = np.full((3, 2), 2)

# print(arr.dot(arr2))

# 배열을 전치해줘야 연산이 가능하다.
print((arr.T).dot(arr2)) # 방법1
print('='*20)
print(arr.dot(arr2.T)) # 방법2
```
